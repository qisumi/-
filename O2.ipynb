```python
# ChatGPT-4
import pandas as pd
from docx import Document

# 安装必要的库（如果还未安装）
# pip install python-docx pandas


def create_custom_document(student, template_path, output_path):
    # 读取Word模板
    doc = Document(template_path)

    # 替换模板中的占位符
    for paragraph in doc.paragraphs:
        for key, value in student.items():
            if f'{{{{ {key} }}}}' in paragraph.text:
                paragraph.text = paragraph.text.replace(
                    f'{{{{ {key} }}}}', str(value))

    # 保存新的Word文档
    doc.save(output_path)


# 读取学生信息的CSV文件
students_df = pd.read_csv('students.csv')

# Word模板文件路径
template_path = 'template.docx'

# 为CSV文件中的每个学生生成一个Word文件
for index, student in students_df.iterrows():
    output_path = f"Student_{student['姓名']}.docx"
    create_custom_document(student, template_path, output_path)

print("所有文档已生成。")
```

```python
# ChatGPT-3.5
import csv
from docx import Document

# 打开CSV文件
csv_file = 'student_data.csv'  # 请将此路径替换为包含学生信息的CSV文件的实际路径

# 打开模板Word文件
template_docx = 'template.docx'  # 请将此路径替换为模板Word文件的实际路径
template = Document(template_docx)

# 读取CSV文件中的学生数据并生成新的Word文档
with open(csv_file, 'r', newline='', encoding='utf-8') as csvfile:
    csv_reader = csv.DictReader(csvfile)
    for row in csv_reader:
        student_name = row['姓名']
        student_age = row['年龄']
        party_application_time = row['入党申请时间']
        class_position = row['班级内职务']

        # 复制模板生成新的Word文档
        new_doc = Document()
        for element in template.element.body:
            new_doc.element.body.append(element)

        # 替换模板中的占位符文本
        for paragraph in new_doc.paragraphs:
            for run in paragraph.runs:
                if '{姓名}' in run.text:
                    run.text = run.text.replace('{姓名}', student_name)
                if '{年龄}' in run.text:
                    run.text = run.text.replace('{年龄}', student_age)
                if '{入党申请时间}' in run.text:
                    run.text = run.text.replace(
                        '{入党申请时间}', party_application_time)
                if '{班级内职务}' in run.text:
                    run.text = run.text.replace('{班级内职务}', class_position)

        # 保存生成的新Word文档
        new_doc_file = f'{student_name}_profile.docx'  # 根据学生姓名生成文件名
        new_doc.save(new_doc_file)

print("新的Word文档已生成。")
```

```python
# Github Copilot
import pandas as pd
from docx import Document

# 读取CSV文件
students = pd.read_csv('students.csv')

# 读取模板Word文件
template = Document('template.docx')

# 遍历每个学生
for index, row in students.iterrows():
    # 创建新的Word文件
    doc = Document()
    # 遍历模板中的每个段落
    for paragraph in template.paragraphs:
        # 替换模板中的变量
        text = paragraph.text.replace('{姓名}', row['姓名'])
        text = text.replace('{年龄}', str(row['年龄']))
        text = text.replace('{入党申请时间}', row['入党申请时间'])
        text = text.replace('{班级内职务}', row['班级内职务'])
        # 添加到新的Word文件中
        doc.add_paragraph(text)
    # 保存新的Word文件
    doc.save(f'{row["姓名"]}.docx')
```

```python
# 文心一言
import pandas as pd
from docx import Document

# 读取CSV文件


def read_csv(filename):
    data = pd.read_csv(filename)
    return data

# 创建Word文档


def create_word_document(template_filename, data):
    doc = Document(template_filename)
    for index, row in data.iterrows():
        # 假设模板Word文件中的占位符是 {name}, {age}, {application_date}, {position}
        doc.add_paragraph(
            f'{row["name"]}, Age: {row["age"]}, Application Date: {row["application_date"]}, Position: {row["position"]}')
    doc.save(f'output_{template_filename}')

# 主函数


def main():
    csv_filename = 'student_data.csv'  # 替换为你的CSV文件名
    template_filename = 'template.docx'  # 替换为你的Word模板文件名
    data = read_csv(csv_filename)
    create_word_document(template_filename, data)


if __name__ == '__main__':
    main()
```
